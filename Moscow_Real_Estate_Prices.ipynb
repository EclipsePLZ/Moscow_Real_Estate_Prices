# Building a simple (base) model

Feature selection. X - input parameters, Y - price of the apartment to predict

```python
df_moscow.columns
```

```python
df_moscow = df_moscow.drop_duplicates(['geo_lat', 'geo_lon', 'building_type', 'level', 'levels',
               'rooms', 'area', 'kitchen_area', 'object_type', 'kremlin_dist'])
```

```python
x = df_moscow[['geo_lat', 'geo_lon', 'building_type', 'level', 'levels',
               'rooms', 'area', 'kitchen_area', 'object_type', 'kremlin_dist']]
y = df_moscow['price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
from sklearn import *
```

Standardizing

```python
numeric_features_x = ['geo_lat', 'geo_lon', 'level', 'levels', 'rooms', 'area',
                      'kitchen_area', 'kremlin_dist']
```

```python
x_stand = x
y_stand = y
```

```python
x_stand[numeric_features_x] = (x_stand[numeric_features_x] - x_stand[numeric_features_x].mean()) / x_stand[numeric_features_x].std()
y_stand = (y_stand - y_stand.mean()) / y_stand.std()
```

```python
x_stand = pd.get_dummies(x_stand)
```

Dividing into test and training samples

```python
X_train, X_test, y_train, y_test = train_test_split(x_stand, y_stand, test_size=0.3, random_state=0)
```

# Building a complex model

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
```

Connecting the GPU

```python
device_name = tf.test.gpu_device_name()
if device_name != '/device:GPU:0':
  raise SystemError('GPU device not found')
print('Found GPU at: {}'.format(device_name))
```

Creating a model

```python
model = Sequential()
model.add(Dense(500, input_shape=(16,), activation='relu'))
model.add(BatchNormalization())
model.add(Dense(200, activation='relu'))
model.add(Dense(200, activation='relu'))
model.add(Dense(1))
model.compile(optimizer='adam',  loss='mse', metrics=['mae'])
```

Training

```python
with tf.device('/device:GPU:0'):
  history = model.fit(X_train, y_train, validation_split=0.2,
                    verbose=1, epochs=300, batch_size=256)
```

```python
plt.figure(figsize=(12,8))
plt.plot(history.history['loss'], label='loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.grid(True)
plt.legend()
plt.show()
```

Prediction for test data

```python
y_predict_MLP = model.predict(X_test)
```

Assessing the quality of the model

```python
metrics.r2_score(y_test, y_predict_MLP)
```

# Clustering

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
X = np.array(x_stand)
```

```python
numClusters = 10
```

Model building

```python
model_kmeans = KMeans(n_clusters=numClusters, init='k-means++', random_state=0).fit(X)
```

Finding centroids

```python
centroids = model_kmeans.cluster_centers_
```

Convert to two-dimensional space

```python
pca = PCA(n_components=2)
```

```python
pca_fit = pca.fit(X)
```

```python
principalComponents = pca_fit.transform(X)
```

```python
principalDF = pd.DataFrame(data=principalComponents,
                           columns=['principal component 1', 'principal component 2'])
```

```python
colors =['red','green','blue','cyan','yellow', 'lime','orange','coral','brown','peru','khaki','tan']
centroidColor= []
for item in range(numClusters):
  centroidColor.append(colors[item])
```

```python
labels = model_kmeans.predict(X)
```

```python
dataPointColor = []
for row in labels:
  dataPointColor.append(colors[row])
```

Let's represent the obtained clusters

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(1,1,1)
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
plt.scatter(principalDF['principal component 1'], principalDF['principal component 2'],
            c=dataPointColor, s=50, alpha=0.5)

centroids_transformed = pca_fit.transform(centroids)
plt.scatter(centroids_transformed[:, 0], centroids_transformed[:, 1],
            c=centroidColor, s=200, marker=('x'))
plt.show()
```
